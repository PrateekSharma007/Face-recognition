```python
import numpy as np 
import cv2 

cap = cv2.VideoCapture(0)  #camera  

#face Detection 
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")


skip = 0 
face_data = [] 
dataset_path = "./data/"
file_name = input("enter the name of the person: ")

while True : 

    ret,frame = cap.read() 

    if(ret == False) : 
        continue 

    gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    
    faces = face_cascade.detectMultiScale(frame , 1.3,5)
    # print(faces)
    faces = sorted(faces , key = lambda f:f[2]*f[3])

    for face in faces[-1:] : #largest face in the end
        x,y,w,h = face 
        cv2.rectangle(frame,(x,y) , (x+w,y+h),(255,0,0),2)

        #get the required face , region of interest 

        offset = 10 
        face_sectionn = frame[y-offset:y+h+offset,x-offset:x+w+offset]

        face_sectionn = cv2.resize(face_sectionn,(100,100))
        skip += 1

        if(skip%10 ==0): 
            face_data.append(face_sectionn)
            print(len(face_data))


    cv2.imshow("Frame" , frame) 
    cv2.imshow("Face section" , face_sectionn)

 

    key_pressed  = cv2.waitKey(1) & 0xFF #8bit integer 
    if(key_pressed == ord('q')) : 
        break 

face_data = np.asarray(face_data)
face_data = face_data.reshape((face_data.shape[0],-1))
print(face_data.shape) 

np.save(dataset_path+file_name+'.npy',face_data)
print("Data saved successfully") 

cap.release() 
cv2.destroyAllWindows()


# dkfskm
```
